```
# import the necessary packages
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Flatten, Conv2D, MaxPooling2D
from tensorflow.keras.callbacks import TensorBoard
from tensorboardcolab import TensorBoardColab, TensorBoardColabCallback
import pickle
from google.colab import drive
```

```
# load the features and labels with pickle
DATA_PATH = "/content/gdrive/My Drive/Colab Notebooks/TensorFlow/IntroDeepLearning/datasets/dogs-cats/" # replace to your path

pickle_in = open(DATA_PATH + "trainX.pickle", "rb")
trainX = pickle.load(pickle_in)

pickle_in = open(DATA_PATH + "trainY.pickle", "rb")
trainY = pickle.load(pickle_in)
```

```
# scale the raw pixel intensities to the range [0, 1]
trainX = trainX / 255.0
```

```
# define the architecture of the network
model = Sequential()
model.add(Conv2D(256, (3, 3), input_shape=trainX.shape[1:]))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(256, (3, 3)))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(64))
model.add(Activation("relu"))
model.add(Dense(1))
model.add(Activation("sigmoid"))
```

```
# make the TensorBoard callback object
tbc = TensorBoardColab()
```

```
# train the model
model.compile(loss="binary_crossentropy",
             optimizer="adam",
             metrics=["accuracy"])
model.fit(trainX, trainY, batch_size=32, 
          epochs=10, 
          validation_split=0.3, 
          callbacks=[TensorBoardColabCallback(tbc)])
```
